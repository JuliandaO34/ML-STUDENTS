```python
# ==========================================
# 📌 2️⃣ Importar librerías necesarias
# ==========================================
import pandas as pd
from pycaret.regression import *

# ==========================================
# 📌 3️⃣ Cargar el dataset
# ==========================================
file_path = "Boston-house-price-data.csv"
df = pd.read_csv(file_path, delimiter=';')

df.head()

# ==========================================
# 📌 4️⃣ Configurar entorno de PyCaret
# ==========================================
# PyCaret configurará todo:
# convierte variables, divide datos, normaliza si hace falta, etc.
reg = setup(
    data=df,
    target='MEDV',
    session_id=123,    # Para reproducibilidad
    train_size=0.8
)

# ==========================================
# 📌 5️⃣ Comparar TODOS los modelos de regresión disponibles
# ==========================================
# Esto entrena TODOS los modelos y muestra métricas clave:
# MAE, MSE, RMSE, R2, RMSLE, MAPE, etc.
best_model = compare_models()

# Muestra el mejor modelo encontrado
print(f"Modelo con mejor desempeño: {best_model}")

# ==========================================
# 📌 6️⃣ Crear el modelo elegido y evaluarlo
# ==========================================
# Si quieres usar el mejor modelo:
final_model = finalize_model(best_model)

# Ver reportes y gráficas interactivas
evaluate_model(final_model)

# ==========================================
# 📌 7️⃣ Hacer predicciones con datos nuevos
# ==========================================
# Crea un diccionario con los datos de entrada
# Cambia estos valores por los que quieras probar:
new_data = pd.DataFrame([{
    'CRIM': 0.1,
    'ZN': 0.0,
    'INDUS': 8.0,
    'CHAS': 0,
    'NOX': 0.5,
    'RM': 6.0,
    'AGE': 65.0,
    'DIS': 4.0,
    'RAD': 4,
    'TAX': 300,
    'PTRATIO': 15.0,
    'B': 390.0,
    'LSTAT': 12.0
}])


```
